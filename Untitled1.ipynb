##### If an pixel has [255,255,255,255] then it is white otherwise black

```python
import matplotlib.pyplot as plt
```

```python
# Forming an array to track if it is white(1) or black(0)
parameters={
    'color1':0,
    'color2':0,
    'color3':0,
    'color4':0
}
ds={
    'd11':d11,
    'd12':d12,
    'd13':d13,
    'd14':d14
    
}
for c in range(4):
    parameters[f'color{c+1}']=np.zeros((64,64))
    for i in range(64):
        for j in range(64):
            if(ds[f'd1{c+1}'][i][j][0]==255 and ds[f'd1{c+1}'][i][j][1]==255 and ds[f'd1{c+1}'][i][j][2]==255 and ds[f'd1{c+1}'][i][j][3]==255):
                parameters[f'color{c+1}'][i][j]+=1
    parameters[f'color{c+1}']=parameters[f'color{c+1}'].reshape(64*64)
colors=['red','yellow','pink','green']
for c in range(4):
    plt.figure(figsize=(3,3))
    points=0
    for i in coordinates:
        x,y=i
        if (parameters[f'color{c+1}'][points]==1):
            plt.scatter(x,y,c=colors[c])
        points+=1
        
```

```python
def distance(k,centres_x,centres_y,coordinates,color):
    dist=np.full((64*64,k),-1)
    for j in range(len(color)):
        if (color[j]==1):
            for i in range(k):
                twist=coordinates-[centres_x[i],centres_y[i]]
                twist=twist**2
                twist=twist.sum(axis=1)
                twist=np.sqrt(twist)
                for b in range(len(twist)):
                    dist[b][i]=twist[b]
    return dist
```

```python
def kmeans(k,color):
    centres_x=np.random.randint(0,64,(1,k))
    centres_y=np.random.randint(0,64,(1,k))
    centres_y=-centres_y
    centres_x=centres_x[0]
    centres_y=centres_y[0]
    assignment= np.array(list(range(k)))
    loss_prev=0
    for u in range(10):
        dist=distance(k,centres_x,centres_y,coordinates,color)
        assigned=np.full(64*64,-1)
        for i in range(len(color)):
            if (color[i]==1):
                assigned[i]=dist[i].argmin()
        for j in assignment:
            indexes = np.array([i for i, value in enumerate(assigned) if value == j])
            if indexes.size == 0:
                centres_x,centres_y,loss_prev=kmeans(k,color)
                return centres_x,centres_y,loss_prev
            x= indexes % 64
            y= -((indexes / 64).astype(int))
            avg_x= x.sum()/len(x)
            avg_y= y.sum()/len(y)
            centres_x[j]=avg_x
            centres_y[j]=avg_y
        dist=distance(k,centres_x,centres_y,coordinates,color)
        loss=dist.min(axis=1).mean()
        if(loss==loss_prev):
            print(loss_prev)
            return centres_x,centres_y,loss_prev
        loss_prev=loss
        
```
